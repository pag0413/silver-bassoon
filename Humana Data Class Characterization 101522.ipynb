```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading full dataset
df = pd.read_csv("2021_Competition_Training.csv", index_col = 0)

df.head()
```

```python
df.set_index("ID", inplace = True)
df.head()
```

```python
df = df.drop(["rx_generic_dist_gpi6_pmpm_ct_t_9-6-3m_b4",'lang_spoken_cd',"mabh_seg","cons_ltmedicr","hedis_dia_hba1c_ge9"], axis = 1)
```

```python
#dividing into features and labels to apply dummies to features and labels separately. Features and Labels will be joined later.
reduced = df.copy() 
x= reduced.loc[:, reduced.columns != 'covid_vaccination']
y = reduced['covid_vaccination']
y = pd.get_dummies(y)
#y = ["1" if i == "vacc" else "0" for i in y]
y = pd.DataFrame(y)
y.head()
```

```python
#Dummies for features
x = pd.get_dummies(x)
x.shape
```

```python
#For some reason got duplicate columns in dummies so had to drop the duplicate, retain the original instance
x_no_dup = x.copy()
x_no_dup = x_no_dup.loc[:,~x_no_dup.columns.duplicated()]
x_no_dup.shape
```

```python
#Loading the Important Features as determined by XGBoost.  Previously saved to csv file.
imp_df=pd.read_csv("Features_Importance_Final2.csv", index_col = 0)
imp_df.head()
```

```python
imp_df.sort_values(by = "Importance", ascending = False)
len(imp_df[imp_df["Importance"]==0])
imp_feat = imp_df[imp_df["Importance"]!=0]

```

```python
imp_feat.shape
```

```python
imp_feat = imp_feat.sort_values(by = "Importance", ascending = False)
imp_feat
```

```python
#Isolating the top 50 features plus lables for further study
top_imp = imp_feat[:51]
top_imp

```

```python
#Reducing the dataset to contain only the top 50 important features:
x_red2 = x_no_dup.copy()
x_red2 = x_red2[top_imp["Feature_name"]]
x_red2.shape
```

```python
#Joining the dataset containing only the top 50 important features and the label
df_red2= x_red2.merge(y, on= "ID", how = "inner")
#df_red2.to_csv("Humana_training_top_features.csv")
```

```python
#Creating a dataset with only the Important Features (245 of them) that include dummies:
x_red = x_no_dup.copy()
x_red = x_red[imp_feat["Feature_name"]]
x_red.shape
```

```python
df_red = x_red.merge(y, on= "ID", how = "inner")
df_red.shape
```

```python
df_red.head()
```

```python

#df_red.to_csv("Humana_training_reduced_features.csv")

```

```python
#Correlation of the top 50 features
top_feat = df_red2.copy()
imp_cor = top_feat.corr() 
imp_cor
```

```python
sns.heatmap(imp_cor)
```

```python
no_vac_feat = imp_cor.copy()
no_vac_feat = imp_cor['no_vacc']
no_vac_feat.shape
```

```python
no_vac_feat_pos = no_vac_feat.copy()
positive = no_vac_feat > 0
no_vac_feat_pos = no_vac_feat_pos[positive]
no_vac_feat_pos.shape
```

```python
no_vac_feat_pos.sort_values(ascending = False)
```

```python
#Correlation with all 245 important features
feat_cor = df_red.copy()
feat_cor = feat_cor.corr() 
feat_cor
```

```python
no_vac_all = feat_cor.copy()
no_vac_all = no_vac_all['no_vacc']
no_vac_all.shape
```

```python
no_vac_all_pos = no_vac_all.copy()
positive = no_vac_all > 0
no_vac_all_pos = no_vac_all_pos[positive]
no_vac_all_pos.shape
```

```python
no_vac_all_pos.sort_values(ascending = False)
```

```python
no_vacc_pos_df = pd.DataFrame(no_vac_all_pos)
no_vacc_pos_df.rename(columns = {"no_vacc": "no_vacc correlation"}, inplace = True)
no_vacc_pos_df.head()
```
