```python
import matplotlib.pyplot as plt
from matplotlib import dates
import matplotlib.lines as mlines
import numpy as np
import seaborn as sns
import pandas as pd
sns.set_style('whitegrid')
%matplotlib inline
```

```python
combined = pd.read_csv('CombinedGME.csv')
combined.rename(columns={'Unnamed: 0': 'Date'}, inplace  = True)
combined['Date'] = pd.to_datetime(combined['Date'])
combined.set_index('Date', inplace = True)
combined.fillna(0, inplace = True)
combined
```

```python
#combined.to_csv('FullGME031821.csv')
```

```python
full=pd.read_csv('FullGME031821.csv')
full['Date'] = pd.to_datetime(full['Date'])
full.set_index('Date', inplace = True)
#full.drop([116,:], inplace  = True)
full.sort_values('Date', axis = 0,ascending = False)

```

```python
#nov20_feb21 = combined[(combined.index >= '2020-11-01') & (combined.index < '2021-02-21')]
```

```python
recent = full.copy()
recent = recent.loc['2020-11-01':'2021-02-20',:]
recent = recent.iloc[:-3,:]
recent.tail(30)
```

```python
blue_line = mlines.Line2D([],[], color = 'blue', label = 'Number of Posts') 
orange_line = mlines.Line2D([],[], color = 'orange', label = 'Trading Volume') 
red_line = mlines.Line2D([],[], color = 'red', label = 'Number of Comments') 
green_line = mlines.Line2D([],[], color = 'green', label = 'Price per Share') 
purple_line = mlines.Line2D([],[], color = 'purple', label = 'Max Author Karma Score')
black_line = mlines.Line2D([],[], color = 'black', label = 'Number of Deleted Authors')
```

```python

fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax1.plot(recent.index, recent['Number of Posts'], label = 'Number of Posts')
ax2.bar(recent.index, recent['Volume'], width = 1,alpha = 0.2, color = 'orange', label = "Volume" )
ax1.set_title('Number of Posts vs GME Trading Volume')
ax1.set_ylabel('Number of Posts')
ax2.set_ylabel('GME Trading Volume (x1e8)')

plt.legend(handles = [blue_line,orange_line], loc = 2);

```

```python
fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()

ax1.plot(recent.index,recent['Number of Posts'])
ax2.bar(recent.index,recent['Close'], width = 1, alpha = 0.2,color = 'green')
ax1.set_title('Number of Posts vs GME Closing Price')
ax1.set_ylabel('Number of Posts')
ax2.set_ylabel('GME Closing Price per Share (USD)')
plt.legend(handles = [blue_line,green_line], loc = 2);
```

```python
fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax1.plot(recent.index, recent['Number of Comments'], color = 'red')
ax2.bar(recent.index, recent['Volume'], width = 1,alpha = 0.2, color = 'orange')
ax1.set_title('Number of Comments vs GME Trading Volume')
ax1.set_ylabel('Number of Posts')
ax2.set_ylabel('GME Trading Volume (x1e8)')
plt.legend(handles = [red_line,orange_line], loc = 2);
```

```python
fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax1.bar(recent.index, recent['Close'], width = 1,alpha = 0.2,color = 'green')
ax2.bar(recent.index,recent['Volume'], width = 1,alpha = 0.2, color = 'orange')
ax1.set_title('GME Closing Price vs GME Trading Volume')
ax1.set_ylabel('GME Closing Price per Share (USD)')
ax2.set_ylabel('GME Trading Volume (x1e8)')
plt.legend(handles = [green_line,orange_line], loc = 2);
```

```python
fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax1.plot(recent.index, recent['Number of Posts'])
ax2.bar(recent.index, recent['Number of Comments'], width = 1,alpha = 0.2, color = 'red')
ax1.set_title('Number of Posts vs Number of Comments')
ax1.set_ylabel('Number of Posts')
ax2.set_ylabel('Number of Comments')

plt.legend(handles = [blue_line,red_line], loc = 2);
```

```python
fig, ax1 = plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax1.plot(recent.index, recent['Max Author Karma Score'], color = 'purple')
ax2.bar(recent.index, recent['Volume'], width = 1,alpha = 0.2, color = 'orange')
ax1.set_title('Max Author Karma Score vs Volume')
ax1.set_ylabel('Max Author Karma Score')
ax2.set_ylabel('GME Trading Volume (x1e8)')

plt.legend(handles = [purple_line,orange_line], loc = 2);
```

```python
fig, ax1= plt.subplots(figsize = (12,8))
ax2 = ax1.twinx()
ax2.plot(recent.index, recent['Deleted Authors'], color = 'black')
ax1.plot(recent.index, recent['Number of Posts'])
ax1.set_title('Number of Deleted Authors vs Number of Posts')
ax1.set_ylabel('Number of Posts')
ax2.set_ylabel('Number of Deleted Authors')


plt.legend(handles = [black_line,blue_line], loc = 2);
```
