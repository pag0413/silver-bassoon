```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Features_Importance_Final2.csv", index_col = 0)
df.head()
```

```python
df_sorted = df.sort_values(by="Importance", ascending = False)
df_sorted.head(25)
```

```python
fig, ax = plt.subplots()
df_sorted["Importance"][:50].plot.bar(ax=ax,figsize = (18,12))
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
ax.set_xticklabels(df_sorted["Feature_name"])
plt.ylim([0.0, 0.075])
fig.tight_layout()
plt.savefig("Most_Important_Features.jpeg")
```

```python
fig, ax = plt.subplots()
df_sorted["Importance"][50:100].plot.bar(ax=ax,figsize = (18,12))
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
ax.set_xticklabels(df_sorted["Feature_name"])
plt.ylim([0.0, 0.075])
fig.tight_layout()
```

```python
fig, ax = plt.subplots()
df_sorted["Importance"][100:150].plot.bar(ax=ax,figsize = (18,12))
ax.set_title("Feature importances using MDI")
ax.set_ylabel("Mean decrease in impurity")
ax.set_xticklabels(df_sorted["Feature_name"])
plt.ylim([0.0, 0.075])
fig.tight_layout()
```
